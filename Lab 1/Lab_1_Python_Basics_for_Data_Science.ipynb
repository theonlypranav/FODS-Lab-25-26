## 2) Lists, tuples, dictionaries

- **list**: ordered, changeable
- **tuple**: ordered, not changeable
- **dict**: key → value mapping


```python
# list
scores = [72, 81, 90, 66]
scores.append(88)
print('scores:', scores)
print('first score:', scores[0])
print('last two:', scores[-2:])

# tuple
shape = (442, 10)
print('shape:', shape)

# dict
student = {'name': 'Asha', 'roll': 17, 'cgpa': 8.2}
print(student)
print('name:', student['name'])
student['cgpa'] = 8.4
print('updated:', student)
```

## 3) If/else and loops

```python
marks = 78
if marks >= 90:
    grade = 'A'
elif marks >= 75:
    grade = 'B'
else:
    grade = 'C'
print('Grade:', grade)

# for loop
nums = [1, 2, 3, 4, 5]
squares = []
for n in nums:
    squares.append(n*n)
print('squares:', squares)

# list comprehension (compact)
squares2 = [n*n for n in nums]
print('squares2:', squares2)

# while loop
k = 3
while k > 0:
    print('k =', k)
    k -= 1
```

## 4) Functions

```python
def mean(values):
    return sum(values)/len(values)

def mse(y_true, y_pred):
    # Mean Squared Error
    return sum((a-b)**2 for a, b in zip(y_true, y_pred)) / len(y_true)

vals = [2, 4, 6, 8]
print('mean:', mean(vals))

y_true = [3, 5, 2]
y_pred = [2.5, 5.5, 1.0]
print('mse:', mse(y_true, y_pred))
```

## 6) file I/O (read/write)
We’ll create a small CSV, save it, and read it back.


```python
import csv

rows = [
    ['area', 'distance', 'price'],
    [250, 5.2, 450],
    [310, 3.8, 520],
    [180, 7.0, 320],
]

with open('mini_apartment.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(rows)

print('Saved mini_apartment.csv')

with open('mini_apartment.csv', 'r') as f:
    print(f.read())
```

## 7) Plot a simple graph
Plotting is needed in most DS labs.
